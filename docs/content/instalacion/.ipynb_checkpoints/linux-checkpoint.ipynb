# INSTALACIÓN LINUX

<p style="text-align: justify;">Polynote actualmente solo se prueba en Linux y MacOS, utilizando el navegador Chrome como cliente</p>

### Instalación:

1. Descargue lo siguiente:

    * [Polynote](https://github.com/polynote/polynote/releases/download/0.2.13/polynote-dist-2.12.tar.gz)

    * [Apache Spark](https://apache.ups.edu.ec/dist/spark/spark-3.0.0-preview2/spark-3.0.0-preview2-bin-hadoop2.7.tgz)
    
    * [Java](https://www.oracle.com/technetwork/java/javase/downloads/jdk8-downloads-2133151.html) (Seleccione según su sistema operativo) o `sudo apt-get install default-jdk` (Esto instala Java11 por defecto).
    

2. Abra el terminal y vaya a la ubicación de descarga:
```bash
cd Users/user_name/Downloads
```

3. Descomprima el archivo tar:
```bash
tar -zxvpf polynote-dist.tar.gz
```

4. Desempaquete el archivo Apache Spark:
```bash
tar -zxvpf spark-2.4.4-bin-hadoop2.7.tgz
cd polynote
```
***Nota:***

    * Cambie `user_name` por su nombre de usuario.
    * Cambie `jdk1.8.0_191.jdk` dentro de la variable de ruta JAVA_HOME con la versión que instaló en su sistema. Para verificar la versión, puede ejecutar `java -version` en una terminal.
    
    ```bash
    export JAVA_HOME=/Library/Java/JavaVirtualMachines/jdk1.8.0_191.jdk/Contents/Home/
    ```
    ```bash
    export SPARK_HOME=/Users/user_name/Downloads/spark-2.4.4-bin-hadoop2.7
    ```
    ```bash
    export PATH="$PATH:$SPARK_HOME/bin:$SPARK_HOME/sbin"
    ```

* Compruebe si spark está configurado correctamente, al ejecutar `spark-submit`, debería ver información de uso.

### Configuración

* <p style="text-align: justify;">Compatibilidad con Python: para ejecutar el código de Python, necesitará tener instalado una versión de <strong>Python 3</strong> y <strong>pip</strong> o <strong>pip3</strong>.</p>

```bash
sudo apt-get install python3.6
```

```bash
sudo apt install python-pip
```

* <p style="text-align: justify;">También tendrá que instalar algunas dependencias de Python como <strong>jep,jedi y virtualenv</strong></p>
			
```bash
pip install jep jedi pyspark virtualenv
```

* <p style="text-align: justify;">Además, es probable que necesite instalar <strong>pandas y numpy</strong></p>

```bash
pip install pandas
```

```bash
pip install numpy
```

### Ejecutar:

```bash
./polynote
```
